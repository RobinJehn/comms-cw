```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Read the CSV file
df = pd.read_csv('goBackN.csv')

# Set the plot size
plt.figure(figsize=(10, 6))

# Define a color palette for three lines
palette = sns.color_palette("gnuplot", n_colors=3)

# Plot each delay column against window size with the palette colors
sns.lineplot(data=df, x='window', y='5ms', marker='o', label='5ms', color=palette[0])
sns.lineplot(data=df, x='window', y='25ms', marker='o', label='25ms', color=palette[1])
sns.lineplot(data=df, x='window', y='100ms', marker='o', label='100ms', color=palette[2])

# Add labels and title
plt.xlabel("Window Size")
plt.ylabel("Throughput (Kbps)")
plt.title("Throughput vs Window Size")

# Set logarithmic scale on x-axis
plt.xscale('log')

# Fix x-axis tick labels to show normal values instead of scientific notation
ax = plt.gca()
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.ticklabel_format(axis='x', style='plain')
ax.set_xticks(df['window'])

plt.legend(title='Delay')
plt.show()
```
